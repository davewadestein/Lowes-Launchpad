# üêß Seaborn Plotting Examples + Custom Themes
* This notebook demonstrates a variety of Seaborn plots
and shows how to customize themes and color palettes

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Use a clean default style
sns.set_theme(style="whitegrid")
```

## 1Ô∏è‚É£ Load Example Data

```python
# if load fails...
# tips = sns.load_dataset("tips")
# penguins = sns.load_dataset("penguins")
# iris = sns.load_dataset("iris")

import pandas as pd

# --- Inline version of tips dataset ---
tips = pd.DataFrame({
    "total_bill": [16.99, 10.34, 21.01, 23.68, 24.59, 25.29, 8.77, 26.88, 15.04, 14.78],
    "tip":        [1.01, 1.66, 3.50, 3.31, 3.61, 4.71, 2.00, 3.12, 1.96, 3.23],
    "sex":        ["Female", "Male", "Male", "Male", "Female", "Male", "Male", "Male", "Female", "Male"],
    "smoker":     ["No", "No", "No", "No", "Yes", "No", "No", "No", "No", "Yes"],
    "day":        ["Sun", "Sun", "Sun", "Sun", "Sun", "Sun", "Sun", "Sun", "Sun", "Sun"],
    "time":       ["Dinner"] * 10,
    "size":       [2, 3, 3, 2, 4, 4, 2, 4, 2, 2]
})

# --- Inline version of penguins dataset ---
penguins = pd.DataFrame({
    "species": ["Adelie", "Adelie", "Chinstrap", "Gentoo", "Gentoo", "Adelie", "Chinstrap", "Gentoo", "Adelie", "Gentoo"],
    "island": ["Torgersen", "Torgersen", "Dream", "Biscoe", "Biscoe", "Dream", "Dream", "Biscoe", "Torgersen", "Biscoe"],
    "bill_length_mm": [39.1, 39.5, 46.5, 50.0, 49.5, 36.7, 45.5, 48.7, 37.3, 46.2],
    "bill_depth_mm":  [18.7, 17.4, 17.9, 16.3, 15.9, 19.0, 17.0, 14.1, 19.6, 14.8],
    "flipper_length_mm": [181, 186, 192, 230, 222, 180, 191, 210, 178, 213],
    "body_mass_g": [3750, 3800, 3800, 5000, 4800, 3450, 3700, 5200, 3600, 5000],
    "sex": ["Male", "Female", "Male", "Female", "Male", "Female", "Male", "Female", "Female", "Male"]
})

# --- Inline version of iris dataset ---
iris = pd.DataFrame({
    "sepal_length": [5.1, 4.9, 6.2, 5.9, 6.7, 5.6, 6.3, 5.0, 5.5, 6.5],
    "sepal_width":  [3.5, 3.0, 2.9, 3.0, 3.1, 2.5, 3.3, 3.6, 2.4, 3.0],
    "petal_length": [1.4, 1.4, 4.3, 5.1, 5.6, 3.9, 6.0, 1.4, 3.8, 5.8],
    "petal_width":  [0.2, 0.2, 1.3, 1.8, 2.4, 1.1, 2.5, 0.2, 1.2, 2.2],
    "species": ["setosa", "setosa", "versicolor", "virginica", "virginica", "versicolor", "virginica", "setosa", "versicolor", "virginica"]
})

tips.head()
```

## 2Ô∏è‚É£ Scatter Plot

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", style="time", s=80)
plt.title("Tip vs Total Bill")
plt.show()
```

## 3Ô∏è‚É£ Histogram

```python
sns.histplot(data=tips, x="total_bill", bins=20, color="skyblue")
plt.title("Distribution of Total Bill")
plt.show()
```

## 4Ô∏è‚É£ Box Plot

```python
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title("Total Bill by Day and Sex")
plt.show()
```

## 5Ô∏è‚É£ Pair Plot

```python
sns.pairplot(iris, hue="species")
plt.suptitle("Iris Pairplot", y=1.02)
plt.show()
```

## 6Ô∏è‚É£ Heatmap

```python
corr = tips.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Heatmap (Tips Dataset)")
plt.show()
```

## 7Ô∏è‚É£ Bar Plot

```python
sns.barplot(data=tips, x="day", y="total_bill", hue="sex", estimator="mean", ci="sd")
plt.title("Average Total Bill by Day and Sex")
plt.show()
```

## 8Ô∏è‚É£ Regression Plot

```python
sns.lmplot(data=tips, x="total_bill", y="tip", hue="sex", aspect=1.3)
plt.title("Regression: Tip vs Total Bill")
plt.show()
```

## 9Ô∏è‚É£ Customizing Themes and Palettes

```python
# ### a) Themes
for theme in ["darkgrid", "whitegrid", "dark", "white", "ticks"]:
    sns.set_theme(style=theme)
    sns.histplot(tips["total_bill"], bins=20, color="royalblue")
    plt.title(f"Theme: {theme}")
    plt.show()
```

```python
# ### b) Color Palettes
for palette in ["deep", "muted", "bright", "pastel", "dark", "colorblind"]:
    sns.set_palette(palette)
    sns.barplot(data=tips, x="day", y="tip", hue="sex")
    plt.title(f"Palette: {palette}")
    plt.show()
```

```python
# ### c) Custom Palette
custom_colors = ["#FF6F61", "#6B5B95", "#88B04B"]
sns.set_palette(custom_colors)

sns.scatterplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="species", s=80)
plt.title("Custom Color Palette Example")
plt.show()
```

```python
# ### d) Continuous Color Map
sns.set_theme(style="whitegrid")
sns.kdeplot(data=tips, x="total_bill", y="tip", fill=True, cmap="viridis")
plt.title("Continuous Color Map (Viridis)")
plt.show()
```

```python
# ### e) Restore Default Theme
sns.set_theme(style="whitegrid", palette="deep")
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title("Restored Default Theme and Palette")
plt.show()
```
